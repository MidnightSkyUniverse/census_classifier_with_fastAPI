## EDA analyses of census data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/census.csv')
```

### Import of data profiling tool for visual analyses of the data

```python
import pandas_profiling
profile = pandas_profiling.ProfileReport(df)
profile.to_widgets()
```

- There is no data that is missing but in 'workclass' there is quite few '?' as a value provided.
- There are whitespaces in both column names and values. We will remove all leading and endng whitespaces

```python
df.columns = df.columns.str.strip()
```

```python
category_columns = [col  for col, dt in df.dtypes.items() if dt == object]
for col in category_columns:
    df[col] = df[col].str.strip()
category_columns
```

```python
numerical_columns = [col  for col, dt in df.dtypes.items() if dt != object]
numerical_columns
```
